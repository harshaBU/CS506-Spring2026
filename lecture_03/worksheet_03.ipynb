### Distance & Similarity

a) In the minkowski distance, describe what the parameters p and d are.

The parameter d describes ther d-dimensional real space the [x_1, ..., x_d] and y = [y_1, ..., y_d] points belong in.  The parameter p describes what power you raise the magnitude of the difference between each x and y value, and then also you take the pth root of the sum of all these magnitude differences to the pth power. 

b) In your own words describe the difference between the Euclidean distance and the Manhattan distance.
The Euclidean Distance follows the Minkowski Distance formula except using p = 2, while the Manhattan Distance uses p = 1.  The Manhattan Distance is the sum of the magnitude differences of the individual components of the y and x vectors, while the euclidian distance is the square root of the sum of the squared differences of the individual components.

Consider A = (0, 0) and B = (1, 1). When:

- p = 1, d(A, B) = 2
- p = 2, d(A, B) = $\sqrt{2} = 1.41$
- p = 3, d(A, B) = $2^{1/3} = 1.26$
- p = 4, d(A, B) = $2^{1/4} = 1.19$

f) what does the jaccard distance account for that the manhattan distance doesn't?
Consider the example of where some binary vector x and another binary vector y (of dimension d) only differ in one of these dimensions.  Then consider binary vectors x', y' each with a dimension of two, and they are completley different.  The manhattan distance between these two vectors in both examples will be both 2.  But obviouslty, the two d-dimension vectors seem to be more similar intuitvely.  Jaccard similarity solves this by considering the ratio between the number of similar components and the number of total components.
